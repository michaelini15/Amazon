```python
# Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read in the dataset

df = pd.read_csv('amazon.csv')
```

```python
pd.set_option('display.max_rows', None) # Allows you to pull through entire dataset
```

```python
# Check out the data

df.head()
```

```python
df.info()
```

```python
# DATA CLEANING
```

```python
# Check for duplicates in each column (1465 - the # shown = # of duplicates)

df.nunique()
```

```python
df_nodups = df.drop_duplicates(subset='about_product') # Drop rows with duplicates in the about_product column
```

```python
df_nodups.nunique()
```

```python
# Check for missing data

df_nodups.isnull().sum()
```

```python
# Add a new column to contain only the first part of the category values, called "more_general_category"

df_nodups['more_general_category'] = df_nodups['category'].apply(lambda category: category.split('|')[0]) # Ignore red
```

```python
df_nodups['category'].nunique() # Unique categories before
```

```python
df_nodups['more_general_category'].nunique() # Unique categories in new column
```

```python
# Add a new column to contain the number of characters for the about_product column, called "description_length"

df_nodups['description_length'] = df_nodups['about_product'].str.len() # Ignore red
```

```python
df_nodups[['category','more_general_category','about_product','description_length']].sort_values('description_length').head()
```

```python
# DATA EXPLORATION
```

```python
# Order the new more_general_category by average rating

df_nodups.groupby(['more_general_category']).mean(['rating']).sort_values('rating', ascending=False)
```

```python
df_nodups[df_nodups['more_general_category']=="OfficeProducts"]['rating'].mean() # Check accuracy
```

```python
# CORRELATION SECTION (data analysis)
```

```python
# Give each company a categorical number (so more_general_category can be checked for correlation)

df_numerized = df_nodups

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(10)
```

```python
# Set up columns to be checked for correlation

df_corrcols = df_nodups[['more_general_category','new_discounted_price','new_actual_price','discount_percentage','rating','new_rating_count','description_length']]

correlation_matrix = df_corrcols.corr()
```

```python
# Create heatmap of correlation_matrix

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix') # There is no correlation of significance (very little correlation at all)
```

```python
# Visualize correlation between rating and descripition_length, as an example

sns.regplot(x='rating',y='description_length',data=df_nodups,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.title('Rating vs Description Length')
plt.xlabel('Rating')
plt.ylabel('Description Length')
```

```python
# Show correlation values in different way (unstacked)

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# Sort the correlation values (ascending)

sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head(10)
```

```python
# Sort and filter the correlation values for significant correlations 

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
